```python
#Dropping "hoa_details" from our dataframe
try:
    zillow_df = zillow_df.drop(["hoa_details"], axis=1)
except Exception as e:
    print("An error occurred:", e)

```

```python
print(zillow_df.columns)
```

```python
# Print all column names as a list
print(zillow_df.columns.tolist())
```

```python
#Many irrelevant data points, keeping only relevant ones.

relevant_cols = ["zpid", "city", "state", "address", "bedrooms", "bathrooms", "price", "yearBuilt", "zipcode", "streetAddress", "livingArea","homeType"]
zillow_df = zillow_df[relevant_cols]
zillow_df.head()
```

```python
num_rows = zillow_df.shape[0]
num_cols = zillow_df.shape[1]
print(f"Number of rows: {num_rows}")
print(f"Number of columns: {num_cols}")
```

```python
#Reducing rows now to only houses with same number of bedrooms and bathrooms to keep pricing consistent
averageBedrooms = 3
averageBathrooms = [2.5, 3.0]

# Filter the DataFrame based on conditions
filtered_df = zillow_df[
    (zillow_df['bedrooms'] == averageBedrooms) &
    (zillow_df['bathrooms'] >= averageBathrooms[0]) &
    (zillow_df['bathrooms'] <= averageBathrooms[1]) &
    (zillow_df['homeType'] == "SINGLE_FAMILY")
]

filtered_df.head()
```

```python
#Determining characteristics of dataframe

num_rows_filtered = filtered_df.shape[0]
num_cols_filtered = filtered_df.shape[1]
print(f"Number of rows: {num_rows_filtered}")
print(f"Number of columns: {num_cols_filtered}")
```

```python
median_living_area = filtered_df["livingArea"].median()
print(median_living_area)
```

```python
lower_percentile = filtered_df['livingArea'].quantile(0.4) 
upper_percentile = filtered_df['livingArea'].quantile(0.6)  
new_filtered_df = filtered_df[(filtered_df['livingArea'] >= lower_percentile) & (filtered_df['livingArea'] <= upper_percentile)]


num_rows_filtered = new_filtered_df.shape[0]
num_cols_filtered = new_filtered_df.shape[1]
print(f"Number of rows: {num_rows_filtered}")
print(f"Number of columns: {num_cols_filtered}")
```

```python
new_filtered_df.head()
```

```python
new_filtered_df = new_filtered_df.sort_values(by=['state', 'city'], ascending=[True, True])

print(new_filtered_df)
```
